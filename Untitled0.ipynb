```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv("operational_excellence on_time_execution_LCL_Not_China.csv")
data_df = pd.DataFrame(data)
pd.set_option('display.max_columns', None)
data_df.head()
```

```python
data_df = data_df.drop(columns=[
                    'Departure port Ocean LOC',
                      'Final port Ocean LOC',
                      'In Full Delivered On Date',
                      'Primary Delay Stage',
                      'Quoted Origin Total Days',
                      'Quoted Port to Port Days',
                      'Quoted Destination Total Days',
                      'Total Quoted Days',
                      'Actual Origin Total Days',
                      'Actual Port to Port Days',
                      'Actual Destination Total Days',
                      'Delta quoted vs actual total days',
                      'Delta actual vs quoted origin days',
                      'Delta actual vs quoted p2p days',
                      'Delta actual vs quoted destination days',
                      'Origin on Time Performance',
                      'Port to Port on Time Performance',
                      'Destination on Time Performance',
                      'OTP',
                      'Shipment ID',
                      'Client ID'])
```

```python
data_df = data_df.dropna()
data_df.head()
```

```python
data_df.info()
```

```python
data_df[['Departure Port Departed On Date',
         'OTP Promised Delivery Date',
         'OTP Promised Delivery Week']] = data_df[['Departure Port Departed On Date',
                                                   'OTP Promised Delivery Date',
                                                   'OTP Promised Delivery Week']].apply(pd.to_datetime)
```

extract year month and day from the date

```python
data_df['Departure_dayofweek'] = data_df['Departure Port Departed On Date'].dt.dayofweek
data_df['Promised_Delive_dayofweek'] = data_df['Departure Port Departed On Date'].dt.dayofweek
data_df['Promised_Delivery_dayofweek'] = data_df['Departure Port Departed On Date'].dt.dayofweek

```

```python
modelling_data_df = data_df.copy()
modelling_data_df = modelling_data_df.drop(columns=[
    "Departure Port Departed On Date",
    "OTP Promised Delivery Date",
    "OTP Promised Delivery Week"
])
modelling_data_df.head(10)
```

```python
cat_columns = ["Ocean type category","Ocean LCL subcategory","Tradelane"]
numerical_columns = ['Departure_dayofweek'	,'Promised_Delive_dayofweek'	,'Promised_Delivery_dayofweek']
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.ensemble import RandomForestRegressor

import numpy as np
# creating dummies for first three columns
X_cat_columns = OneHotEncoder(sparse_output=False, handle_unknown='ignore').fit_transform(modelling_data_df[cat_columns])

# Label encode the high-cardinality column
X_port_pair = LabelEncoder().fit_transform(modelling_data_df["Port pair"]).reshape(-1, 1)

# Convert multiple numerical columns to numpy array
X_num = modelling_data_df[numerical_columns].to_numpy()

# Concatenate all parts
X = np.concatenate([X_cat_columns, X_port_pair, X_num], axis=1)

# Get target variable
y = modelling_data_df['Total Actual Days'].to_numpy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
rf_model = RandomForestRegressor(n_estimators=500, random_state=0)
rf_model.fit(X_train_scaled, y_train)
```

```python
predictions = rf_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,6))
plt.scatter(y_test, predictions, alpha=0.5)
plt.xlabel("Actual Total Days")
plt.ylabel("Predicted Total Days")
plt.title("Actual vs Predicted")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.show()
```
